```python
#prime number check
n=int(input())
if n>1:
    for i in range(2,(n//2)+1):
        if n%i==0:
            print("no")
            break
    else:
        print("yes")
else:
    print("no")
```

```python
#check between range prime numbers
s=int(input())
e=int(input())
l=[]
for i in range(s,e+1):
    if i>1:
       for j in range(2,(i//2)+1):
           if i%j==0:
               break
       else:
           l.append(i)
print(l)
```

```python
#break continue pass
for i in range(0,10,1):
    if i==5:
        break     #if will met condition it will come out from loop
    print(i)
for n in range(0,10,1):
    if n==5:
        continue #if will met condition it skip current iteration
    print(n)
```
