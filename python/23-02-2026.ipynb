```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
months=['jan','feb','mar','apr','may','jun','jul','aug']
traff=[150,200,180,90,30,90,70,95]
#create a line plot
plt.plot(months,traff)
#add lable and tiltle to
plt.xlabel('month')
plt.ylabel('monthlytariff')
plt.title('monthly tariff titld')
#create the line plot with custom apperance
plt.plot(months,traff,marker='^',markersize=30,color='green',linestyle='-',linewidth=3)
#display the plot
plt.show()
```

```python
#multiple line plot
months=['jan','feb','mar','apr','may','jun','jul','aug']
a=[35,40,50,55,60,59,29,35]
b=[35,75,90,30,45,55,85,95]
#create a line plot for product a with a blue line and circular markers
plt.plot(months,a,marker='o',linestyle='--',color='blue',label='producta')
#create a line plot for product b with a red dashed and square marker
plt.plot(months,b,marker='s',linestyle='-',color='red',label='productb')
#add label and title to it so
plt.xlabel('month')
plt.ylabel('monthly revenue comparision in($100)')
plt.title('monthly revenue comparision')
#display a legend to difference between a and b
plt.legend()
#display the plo
plt.show()
```

```python
#stock market data
month=['jan','feb','mar','apr','may','jun','jul','aug']
companya=[98,99,35,78,85,95,35,55]
companyb=[35,79,55,99,75,37,91,73]
#create line plot for product a
plt.plot(month,companya,marker='o',linestyle='-',color='orange',label='company-a')
#create line plot for company b
plt.plot(month,companyb,marker='s',linestyle='--',color='violet',label='company-b')
plt.legend()
plt.xlabel('months')
plt.ylabel('revenue of company monthly wise ar to')
#saving the plot as an image to
plt.savefig('C:/Users/Administrator/Downloads/piechart1')
plt.show()
```
