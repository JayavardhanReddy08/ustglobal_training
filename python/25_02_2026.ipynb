```python
#Thalassemia vs Disease
#Cross-tabulate thal and target
crosstab = pd.crosstab(data['thal'],data['condition'])
crosstab
```

```python
#conver to percentage
crosstab_pct=crosstab.div(crosstab.sum(axis=1), axis=0) * 100
crosstab
```

```python
#plot stach barchat
crosstab_pct.plot(kind='bar', stacked=True, figsize=(8,6), color=['blue','red'])
plt.xlabel('Thalassemia')
plt.ylabel('Percentage')
plt.title('Disease Presence by Thalassemia Type')
plt.legend()
plt.show()
```

```python
#multi factor risk analysis on
age=data['age'].values
chol=data['chol'].values
bp=data['trestbps'].values
condition=data['condition'].values
age
chol
bp
condition
m1=(age > 50) & (chol > 240) & (bp > 140)
subset_targets=condition[m1]
disease_percentage=(np.sum(subset_targets== 1)/len(subset_targets))*100
disease_percentage
```

```python
#create risk score
data['risk_score'] = (data['chol'] / 200) + (data['trestbps'] / 120) + data['oldpeak']
def classify_risk(score):
    if score < 3:
        return 'Low Risk'
    elif score < 5:
        return 'Medium Risk'
    else:
        return 'High Risk'
data['risk_category'] = data['risk_score'].apply(classify_risk)
data['risk_category'].value_counts()
plt.figure(figsize=(8,6))
data['risk_category'].value_counts().plot(kind='bar', color=['green','orange','red'])
plt.xlabel('Risk Category')
plt.ylabel('Number of Patients')
plt.title('Distribution of Custom Risk Score Categories')
plt.show()
```
