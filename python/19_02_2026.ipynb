```python
def get_sum(lst):
    """
    this function returns sum of elements
    """
    s=0
    for i in range(0,len(lst)):
        s=s+lst[i]
    return s
l=get_sum([1,2,3,4,5,6,7,8,9,7,3,9])
print(l)
        
```

```python
def fac(n):
    """
    this function for factiorial
    """
    f=1
    if n>1:
        for i in range(2,n+1):
            f=f*i
    return f
l=fac(2)
print(l)
```

```python
import math
def fac(n):
    """
    factorial using functio
    """
    l=math.factorial(n)
    return l
l=fac(2)
print(l)
```

```python
#inheritance
class sports:
    def __init__(self,model,color,cost):
        self.__model=model
        self.__color=color
        self.__cost=cost
    def setmodel(self,modelname):
        self.__model=modelname
    def getmodel(self):
        return self.__model
    def setcolor(self,colorname):
        self.__color=colorname
    def getcolor(self):
        return self.__color
    def setcost(self,costnam):
        self.__cost=costnam
    def getcost(self):
        return self.__cost
class shoe(sports):
    def __init__(self,model,color,cost,brand):
        super().__init__(model,color,cost)
        self.__brand=brand
sh1=shoe("shoes","black",2333,"alo")
print(sh1.getmodel())
```
