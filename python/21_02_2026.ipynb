```python
#iterator
class NumberMy:
    def __iter__(self):
        self.num=1
        return self
    def __next__(self):
        if self.num<=5:
            x=self.num
            self.num+=1
            return x
        else:
            raise StopIteration
mynumber=NumberMy()
myiter=iter(mynumber)
for i in myiter:
    print(i)
```

```python
class even:
    def __iter__(self):
        self.num=1
        return self
    def __next__(self):
        if self.num<=10:
            if self.num%2==0:
                x=self.num
                self.num+=1
                return x
            else:
                self.num+=1
        else:
            raise StopIteration
evennum=even()
nume1=iter(evennum)
for i in nume1:
    print(i)
                      
```

```python
#generator
def bmr_generator():
    for i in range(1,6):
        yield i
gen=bmr_generator()
for value in gen:
    print(value)
```

```python
def bmr_generator():
    for i in range(1,7):
        if i%2!=0:
            yield i
gen=bmr_generator()
for value in gen:
    print(value)
```

```python
def bmr_generator(*n):
    n=n
    for i in n:
        if len(i)<=3:
            yield i
gen=bmr_generator("ili","divya","ram","raj","lal")
for divya in gen:
    print(divya)
```
