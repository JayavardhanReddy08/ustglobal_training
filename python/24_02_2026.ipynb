```python
#sample data for stack plot
q=['q1','q2','q3','a4','q5']
e=[10000,15000,13000,11000,9000]
c=[5000,7000,9000,3000,10000]
h=[9000,3000,1000,5000,3000]
#create scatter plot
plt.figure(figsize=(10,6))# set the figure size width and height
plt.stackplot(q,e,c,h,labels=['electronics','clothing','homeapliences'])
#add labels and title legend
plt.xlabel('quarters')
plt.ylabel('sales($)')
plt.title('product category sale')
plt.legend()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
#generate data fro 3d axes
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
x,y=np.meshgrid(x,y)
z=np.sin(np.sqrt(x**2 + y**2))
#3D scatter plot
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x,y,z,c=z,cmap='viridis')
ax.set_xlabel('xaxis')
ax.set_ylabel('ylabel')
ax.set_zlabel('zaxis')
plt.show()
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
y4=np.exp(-x)
#create 2x2 sub plot to
fig,axs=plt.subplots(2,2,figsize=(10,8))
#plot data each on subplot
axs[0,0].plot(x,y1,'r')#top left subplot
axs[0,0].set_title('sin_wav')
axs[0,1].plot(x,y2,'b')#top right subplot
axs[0,1].set_title('cosin_wav')
axs[1,0].plot(x,y3,'g')#bottom left subplot
axs[1,0].set_title('tanget_lin')
axs[1,1].plot(x,y4,'m')#bottom right subplot
axs[1,1].set_title('expotentional delay plot')
#adjust layout to
plt.tight_layout()
plt.show()
```
